```python
# Create dependencies for the data set
import pandas as pd
import numpy as np
import datetime
```

```python
#read in the csv file and set the index of 'Date'
credit_card = pd.read_csv('Data/CC Report.csv', index_col='Gift Date', parse_dates=True)
```

```python
# Display contents of data file
credit_card
```

```python
# Remove columns that we do not need
credit_card.iloc[:, [1,2,6,7,8,9,10,11,12,14,15,16]]
```

```python
# Replace NaN values from dataframe
credit_card.fillna("", inplace=True)
```

```python
credit_card.head()
```

```python
# Sort the columns in order desired
credit_card_sort = credit_card[['Gift Type', 'Pay Method', 'Gift Subtype', 'Fund Description', 'Name', 'Gift Amount', 'Soft Credit Recipient', 'Appeal', 'Package Description', 'Campaign', 'Sponsorship/Leadership', 'Reference']]
credit_card_sort
```

```python
group_credit_card = credit_card_sort.groupby(['Fund Description'])
group_credit_card.head()
```
